```python
for i in df['fname']:
    tsv = pd.read_csv(i, sep = '\t')
    csv = pd.DataFrame()
    csv['lat'] = tsv['latitude']
    csv['lon'] = tsv['longitude']
    csv['alt'] = tsv['altitude_meters']
    csv['timestamp'] = tsv['timestamp']
    csv.to_csv('./csv_files/' + i.split('/')[-1].split('.')[0] +'.csv')
```
